```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
import pandas as pd
import numpy as np
from scipy.stats import norm
```

```python
#labels = ['unknown', 'prices', '']
real_estate_prices = pd.read_csv("./houses.csv")
rep_values = real_estate_prices.iloc[:,1]
rep_mean = rep_values.mean()
rep_median = rep_values.median()
rep_std = rep_values.std()
rep_min = rep_values.min()
rep_max = rep_values.max()
#real_estate_prices
print(f'{rep_mean}, {rep_median}, {rep_std}, {rep_min}, {rep_max}')
#real_estate_prices
rep_values.to_numpy()
```

```python
# Plot of the distribution of the prices
# The plot is initially weird looking because the values are discrete, 
# if we convert them to "continuous" values, the histogram will be a good visualisation tool

n_bins = rep_values.size
#bins = [i for i in range(0, rep_max, int(rep_max/6))]
#print(rep_values.values.size)
rep_valuesf = rep_values.values.astype(np.float32)/256
plt.hist(rep_values, bins=np.arange(rep_min, rep_max), log=True)
plt.ylabel('Frequency')
plt.xlabel('Price')
plt.title('Prices of real estates in England')
plt.show()
```

```python
# Plots (histograms) for real estate prices in and outside of London
rep_label = real_estate_prices.iloc[:,13]
cond = rep_label.str.contains('LONDON')
rep_London = real_estate_prices[cond]
#rep_London

mask = [not c for c in cond]
rep_outside_London = real_estate_prices[mask]
#rep_outside_London

fig, axes = plt.subplots(1,2,figsize=(12,4))
axes[0].hist(rep_London.iloc[:,1].values, bins=rep_London.size)
axes[0].set_title("Real estate prices in London")

axes[1].hist(rep_outside_London.iloc[:,1].values, bins=rep_outside_London.size)
axes[1].set_title("Real estate price outside of London")
plt.show()
```

```python
# Testing plotting as gaussian distribution
x_axis = rep_values.values #np.arange(int(rep_min), int(rep_max), 0.5) # np.arange(rep_min, rep_max)
plt.plot(x_axis, norm.pdf(x_axis, rep_mean, rep_std))
plt.show()
```
